```python
import folium
import geopandas as gpd
import matplotlib.pyplot as plt
import osmnx as ox
import pandas as pd
ox.config(use_cache=True)
```

```python
qm_latlng = ox.geocode('Queen Mary Hospital, Hong Kong')
qm_latlng
```

```python
pw_latlng= ox.geocode('Prince of Wales Hospital, Hong Kong')
pw_latlng
```

```python
#explicit plot map
#m=folium.Map([22.16,114.0752],zoom_start=4)
#alternative plot
m=folium.Map(location=qm_latlng,zoom_start=10)
folium.Marker(qm_latlng).add_to(m)
folium.CircleMarker(
    location=qm_latlng,
    radius=10,
    fill=True,
    popup=folium.Popup('Queen Mary Hospital')
).add_to(m)
folium.CircleMarker(location=pw_latlng,popup=folium.Popup('Prince of Wales Hospital')).add_to(m)
m.save('webmap/hongkong.html')
m
```
