```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Image
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
# configure the inline image display
img_folder = 'images'
extension = 'png'
size = 240
dpi = 40
```

```python
# geocode MMC to lat-lng
Melb = ox.geocode('Melbourne, Victoria, Australia')
Sydney = ox.geocode('100 market st, Sydney,NSW, Australia')
Canberra = ox.geocode('parliament drive, Canberra, ACT, Australia')
Brisbane = ox.geocode('Brisbane, QLD, Australia')
Boston = ox.geocode('Boston, MA, USA')
Providence = ox.geocode('Providence, Rhode Island, USA')

Melb
Sydney
Canberra
Brisbane
```

```python
place = 'Melbourne'
#point = (37.793897, -122.402189)
point=Melb
fig, ax = ox.plot_figure_ground(point=point, filename=place, dpi=dpi)
Image('{}/{}.{}'.format(img_folder, place, extension), height=size, width=size)
```
