```python
#topic model using NMF
import pandas as pd
#import gensim
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
```

```python
#error with utf-8
#resolve add comment encoding =....
#data from comments post on Youtube Stroke heroes act fast #YHzz2cXBlGk
docs=pd.read_csv('../../Rcourse/SocialMedia2018/Utube/Feb_01_1_49_59 PM_2018_AEDT_YoutubeData.csv',encoding = 'unicode_escape', engine ='python')
docs.head()
```

```python
# use tfidf by removing tokens that don't appear in at least 50 documents
vect = TfidfVectorizer(min_df=2, stop_words='english')
# Fit and transform
X = vect.fit_transform(docs.Comment)
```

```python
# Create an NMF instance: model
# the 10 components will be the topics
model = NMF(n_components=10, random_state=5)
# Fit the model to TF-IDF
model.fit(X)
# Transform the TF-IDF: nmf_features
nmf_features = model.transform(X)
```

```python
# Create a DataFrame: components_df
components_df = pd.DataFrame(model.components_, columns=vect.get_feature_names())
components_df
```

```python
for topic in range(components_df.shape[0]):
    tmp = components_df.iloc[topic]
    print(f'For topic {topic+1} the words with the highest value are:')
    print(tmp.nlargest(10))
    print('\n')
```
